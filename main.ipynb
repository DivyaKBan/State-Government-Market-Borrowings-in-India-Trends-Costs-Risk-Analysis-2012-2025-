```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the raw dataset 

raw_df=pd.read_excel("Details of State Governments Market Borrowings - Copy.xlsx")
raw_df.head()
```

```python
# Loading the dataset while removing the first 4-5 empty or meaning-less rows and the first empty column

df=pd.read_excel("Details of State Governments Market Borrowings - Copy.xlsx",skiprows=5, usecols=lambda col: col != df.columns[0])
df.head()
```

```python
df.columns
```

```python
# Renaming the columns

df = df.rename(columns={
    'Date of': 'auction_date',
    'Unnamed: 2': 'issue_date',
    'Unnamed: 3': 'maturity_date',
    'Maturity':'maturity_residual_period',
    'Unnamed: 6':'maturity_years',
    'Bids Received':'BR_comp_number',
    'Unnamed: 8':'BR_comp_value',
    'Unnamed: 9':'BR_non-comp_number',
    'Unnamed: 10':'BR_non-comp_value',
    'Bids Accepted':'BA_comp_number',
    'Unnamed: 12':'BA_comp_value',
    'Unnamed: 13':'BA_non-comp_number',
    'Unnamed: 14':'BA_non-comp_value',
    'Name of the State': 'state',
    'Notified Amount(Rupees Crores)': 'notified_amount',
    'Amount Raised (₹ Crore)': 'amount_raised',
    'Weighted Avg. Yield': 'avg_yield'
})
df.columns
```

```python
df.head()
```

```python
df = df.drop(df.index[:4])
```

```python
# Dropping the empty columns
df = df.drop(columns=["Devolvement on Primary Dealers (Rupees Crores)", "Devolvement/Private placement on RBI (Rupees Crores)"])
```

```python
# Dropping the rows where auction date is not present

df = df.dropna(subset=['auction_date'])
```

```python
df.head()
```

```python
Dates_cols=[
    'auction_date',
    'issue_date',
    'maturity_date',
    'maturity_years'
]

for dates in Dates_cols:
    df[dates]=pd.to_datetime(df[dates], errors='coerce')
# df['auction_date'] = pd.to_datetime(df['auction_date'], errors='coerce')

numeric_cols = [
    'notified_amount',
    'avg_yield',
]

for col in numeric_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
df.head()
```

```python
df['coupon_rate'] = (
    df['Nomenclature Loan']
    .str.extract(r'(\d+\.?\d*)%')
    .astype(float)
)
df['state'] = (
    df['Nomenclature Loan']
    .str.extract(r'%\s([A-Z]+)')
)
state_map = {
    'MADHYAPRADESH': 'MADHYA PRADESH',
    'TAMILNADU': 'TAMIL NADU',
    'UTTARAKHAND': 'UTTARAKHAND',
    'ANDHRA': 'ANDHRA PRADESH'
}

df['state'] = df['state'].replace(state_map)

df['instrument'] = (
    df['Nomenclature Loan']
    .str.extract(r'\b(SGS)\b')
)
```

```python
df = df.drop_duplicates()
```

```python
df['year'] = df['auction_date'].dt.year
df['quarter'] = df['auction_date'].dt.to_period('Q')
df = df.drop(columns=['maturity_years'])
df.head()
```
